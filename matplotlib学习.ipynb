# 常用方法
1. plot(*args, **kwargs)用于绘制线或者点
2. title(s, *args, **kwargs)为当前坐标系设置标题
3. show(*args, **kw)展示图片
4. xlabel(s, *args, **kwargs) 设置x轴标签
5. ylabel(s, *args, **kwargs)设置y轴标签
6. bar(*args, **kwargs)绘制条形图
7. scatter(x, y[, s, c, marker, cmap, norm, …])绘制散点图
8. xticks(*args, **kwargs)为x轴的位置设置标记
9. yticks(*args, **kwargs)为y轴位置设置标记
[https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html]

### 调整subplot(子图)周围的间距
subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=None, hspace=None)

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0.1)
```

### 颜色、标记和线型

```python
from numpy.random import randn
fig = plt.figure()
plt.plot(randn(30).cumsum(), 'ko--')
plt.plot(randn(30).cumsum(), color='k', linestyle='dashed', marker='o')
```

##  添加图例

```python
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
